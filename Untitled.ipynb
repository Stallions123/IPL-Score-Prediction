```python
import pandas as pd
import pickle
```

```python
df=pd.read_csv("ipl.csv")
```

```python
df.head()
```

```python
columns_to_remove=['mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels=columns_to_remove,axis=1,inplace=True)
```

```python
df['bat_team'].unique()
```

```python
consistent_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab','Royal Challengers Bangalore', 'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
df.head()
```

```python
df=df[df['overs']>=5.0]
```

```python
df.head()
```

```python
from datetime import datetime
```

```python

df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%d-%m-%Y'))
```

```python
encoded_df=pd.get_dummies(data=df, columns=['bat_team' , 'bowl_team'])
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
encoded_df=encoded_df[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore','bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
       'total']]
```

```python
encoded_df.head()
```
